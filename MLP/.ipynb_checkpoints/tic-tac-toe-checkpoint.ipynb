```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
df = pd.read_csv('tic-tac-toe.data')
df.head()
```

```python
df = df.replace('x', 1).replace('o',2).replace('b',0).replace('positive',1).replace('negative',0)
df.head()
```

```python
list(df.loc[0])
```

```python
def print_game(game):
    print(game[-1])
    for i in range(0,9,3):
        print(game[i], game[i+1], game[i+2])
    print()
    
for game in df.values[:3]:
    print_game(game)
    
```
